# Q1: How have automobile sales changed over the years, and what impact did recessions have on them?

```python
# Extract the year from the 'Date' column and create a new column 'Year'
df['Year'] = df['Date'].dt.year

# Group the data by 'Year' and calculate the sum of 'Automobile_Sales'
yearly_sales = df.groupby('Year')['Automobile_Sales'].sum().reset_index()

# Print the resulting DataFrame to see the sum of automobile sales by year
print(yearly_sales)
```

```python
# Group data by 'Year' and get the sum of 'Automobile_Sales'
yearly_sales = df.groupby('Year')['Automobile_Sales'].sum().reset_index()

# Identify recession years
recession_years = df[df['Recession'] == 'Yes']['Year'].unique()

plt.figure(figsize=(14, 7))
plt.plot(yearly_sales['Year'], yearly_sales['Automobile_Sales'], marker='o')

# Highlight recession years
for year in recession_years:
    plt.axvline(x=year, color='r', linestyle='--', alpha=0.7)

plt.title('Trend of Automobile Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Total Automobile Sales')
plt.grid(True)
plt.show()
```

Observations about the trends and correlations: 
• There are sharp increases and decreases (volatility) rather than a smooth trend, which could indicate that automobile sales are highly sensitive to external factors. 
• There appears to be a cyclical pattern, with peaks and troughs occurring at regular intervals. This could suggest that the automobile sales cycle is correlated with economic cycles, consumer confidence, or other industry-specific factors. 
• Despite the volatility, there seems to be a general upward trend in sales from the early 1980s until the early 2000s. This could indicate overall industry growth during this period, possibly due to economic expansion, technological advancements, or population growth leading to higher demand. 
• After the early 2000s, there is a notable period of decline. This might correspond to specific economic events (such as the 2008 financial crisis) that led to a downturn in consumer spending and confidence. 
• Post the decline, there is a recovery, which is marked by fluctuations which might indicate instability in the market or changes in consumer behavior. 
• There is no consistent linear correlation over time; sales do not consistently increase or decrease with each passing year. Instead, they are influenced by other variables and conditions that change over time. 
• The trends could be influenced by external factors such as economic conditions (recessions, booms).

# Q2. How does seasonality affect automobile sales?

```python
# Analyzing the Effect of Seasonality on Sales

# Extracting month from the 'Date' column
df['Month'] = df['Date'].dt.month

# Grouping data by Month and calculating average Automobile Sales
monthly_sales = df.groupby('Month')['Automobile_Sales'].mean().reset_index()

# Creating a bar chart to show the effect of seasonality
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Automobile_Sales', hue='Month', data=monthly_sales, palette='Set3', dodge=False, legend=False)
plt.title('Effect of Seasonality on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Average Automobile Sales')
plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

The bar chart provided illustrates the average automobile sales by month, allowing us to analyze the effect of seasonality on automobile sales.

Observations: 
• The months with the tallest bars represent the highest average sales, indicating peak purchasing times, while the shorter bars indicate lower sales periods. 
• There seems to be a significant increase in sales towards the end of the year, with December showing one of the highest average sales. This could be due to year-end promotions.
• The middle months, typically spring and summer, also show robust sales figures. This might be due to better weather conditions, which are conducive to car shopping and test-driving activities. 
• There is a noticeable dip in sales at the beginning of the year, with January and February showing lower averages. This could be due to post-holiday financial recovery for consumers or less inclination to purchase vehicles during winter months in many regions.

Recommendations: 
• The chart could help businesses in the automobile industry to strategize their marketing efforts, aligning promotions and advertisements with the high and low sales months. 
• Dealerships might also use this information to manage inventory levels and staffing, preparing for the expected surge or slowdown in sales throughout the year.

Observations:

• Despite varying levels of competition, average automobile sales remain relatively stable. This could suggest that factors other than competition, such as brand loyalty, product quality, or market saturation, may have a more significant impact on sales figures. 
• The chart does not show peak sales at the lowest or highest levels of competition, implying that there isn't a straightforward relationship where less competition equals more sales or vice versa. 
• There seems to be a slight increase in sales when the competition level is at 7, suggesting there could be an optimal level of competition that maximizes sales.

Recommendations:

# Q6. Did recession periods significantly impact automobile sales?

Compare average sales during recession and non-recession years using a bar chart.

```python
# Analyzing the Impact of Recession on Automobile Sales

# Grouping data by Recession status and calculating average Automobile Sales
recession_sales = df.groupby('Recession')['Automobile_Sales'].mean().reset_index()

# Define recession criteria (use a 'Recession' column with 1 for recession years and 0 for no recession years)
# Converting the 'Recession' column to a more readable format
recession_sales['Recession'] = recession_sales['Recession'].map({0: 'No Recession', 1: 'Recession'})

# Creating a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Recession', y='Automobile_Sales', data=recession_sales)
plt.title('Impact of Recession on Automobile Sales')
plt.xlabel('Recession Status')
plt.ylabel('Average Automobile Sales')
plt.show()
```

The bar chart illustrates the impact of recession on automobile sales. It compares the average automobile sales during recession years to those during non-recession years. This visualization helps in understanding how economic downturns, indicated by recessions, might affect the automobile industry's sales.

Observations: 
• There is a noticeable decrease in the average automobile sales during recession periods compared to non-recession periods. 
• The automobile industry appears to be sensitive to economic downturns. Sales are significantly higher when the economy is not in recession. 
• While many factors can influence automobile sales, the graph strongly suggests that recession is a major factor affecting consumer behavior and industry sales.

Recommendations: 
• Understanding the impact of recessions on automobile sales can inform industry and government policy. During recessionary times, interventions such as incentives for buyers or support for manufacturers might be necessary to stabilize the industry.

# Q9. How do price trends correlate with GDP and unemployment rates?

```python
# Setting Seaborn style
sns.set(style="whitegrid")

# Grouping the data by year and calculating mean values for price, GDP, and unemployment rate
yearly_data = df.groupby('Year').agg({'Price': 'mean', 'GDP': 'mean', 'unemployment_rate': 'mean'})

# Creating a figure and a primary axis
plt.figure(figsize=(12, 6))
ax1 = plt.gca()  # Get current axis

# Plotting Average Price with Seaborn (with Matplotlib's axis)
sns.lineplot(data=yearly_data, x=yearly_data.index, y='Price', ax=ax1, color='blue', label='Average Price')
ax1.set_xlabel('Year')
ax1.set_ylabel('Average Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Creating a secondary y-axis for GDP
ax2 = ax1.twinx()
sns.lineplot(data=yearly_data, x=yearly_data.index, y='GDP', ax=ax2, color='green', label='GDP')
ax2.set_ylabel('GDP', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Creating a third y-axis for Unemployment Rate
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # Offset the right spine of ax3
sns.lineplot(data=yearly_data, x=yearly_data.index, y='unemployment_rate', ax=ax3, color='red', label='Unemployment Rate')
ax3.set_ylabel('Unemployment Rate', color='red')
ax3.tick_params(axis='y', labelcolor='red')

# Adding a title and adjusting legend
plt.title('Price Trends vs GDP and Unemployment Rate Over Years')
ax1.legend(loc='upper left')
ax2.legend(loc='lower left')
ax3.legend(loc='upper right')

plt.show()

# Returning the prepared yearly_data for the multi-axis plot
yearly_data.head()
```

The multi-axis line chart and data table illustrate the relationship between average automobile prices, GDP, and the unemployment rate over several years.

Observations: 
• The average price of automobiles (blue line) and GDP (green line) occasionally show parallel movements. For example, from the early 1980s to the early 1990s, both the average price and GDP show upward and downward trends, suggesting that there might be some level of correlation between economic output and automobile pricing during this period. 
• The unemployment rate (red line) generally appears to have an inverse relationship with both GDP and average prices, particularly noticeable in the 1980s and early 1990s. When the unemployment rate is high, GDP tends to be lower, and average automobile prices also tend to decrease, which is consistent with economic theory. 
• There are periods where the trends diverge. For example, there are instances when the average price of automobiles rises while GDP falls, or vice versa. Similarly, the unemployment rate does not always move inversely to the price and GDP lines. 
• In the 2000s, the correlation between these three indicators seems to weaken. The average price of automobiles does not follow a clear trend in response to changes in GDP or the unemployment rate, indicating a potential decoupling of automobile prices from these traditional economic indicators.

Recommendations: 
• During periods of high unemployment, automobile companies might need to implement more aggressive sales strategies, such as offering discounts or more affordable vehicle models, to maintain sales volumes. 
• In times of economic growth, as indicated by rising GDP, companies could focus on higher-end models and emphasize value and quality, which consumers might be more willing to pay for when economic conditions are favorable. 
• Companies should conduct a more detailed analysis of the factors influencing automobile prices, considering elements such as consumer purchasing power, credit availability, and broader economic indicators.

# Q10. How do automobile sales correlate with various economic indicators and advertising expenditures across different months and cities?

```python
# Calculate the correlation matrix of the relevant numerical columns
correlation_columns = ['Consumer_Confidence', 'Advertising_Expenditure', 'GDP', 'unemployment_rate', 'Automobile_Sales']
correlation_matrix = df[correlation_columns].corr()

# Creating the heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap of Correlation Between Automobile Sales and Economic Indicators')
plt.show()
```

Observations: 
• There are no strong correlations between the variables. Most values are close to zero, suggesting a weak relationship. 
• There is a slight negative correlation between 'unemployment_rate' and 'Automobile_Sales' (-0.40), which may imply that higher unemployment could be associated with lower automobile sales. 
• There is a very week positive correlation between 'GDP' and 'Automobile_Sales' (0.01). This could indicate that as the general economy grows, automobile sales could increase. 
• The correlation between 'Advertising_Expenditure' and 'Automobile_Sales' is almost non-existent (-0.01), suggesting that the amount spent on advertising is not directly influencing sales within this dataset. 
• Consumer_Confidence' seems to have very little correlation with 'Automobile_Sales' (0.04), indicating other factors may be more influential in determining sales.

Recommendations: 
• Since there is no apparent correlation between 'Advertising_Expenditure' and sales, it may be beneficial to review current advertising strategies to ensure they are effective and providing a good return on investment.
